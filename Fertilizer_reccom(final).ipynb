```python
# Assuming df_fertilizer is the name of your DataFrame
X = df_fertilizer[['Temparature', 'Humidity ', 'Moisture', 'Soil Type', 'Crop Type', 'Nitrogen', 'Potassium', 'Phosphorous']]
Y = df_fertilizer['Fertilizer Name']
```

```python
# Encode features (X)
X = pd.get_dummies(X, columns=['Soil Type', 'Crop Type'], drop_first=True)

# Encode target (Y)
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```python
# Split the data into training and testing sets (e.g., 80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train the model
model = DecisionTreeClassifier(random_state=2)
model.fit(X_train, Y_train)

# Make predictions
Y_pred = model.predict(X_test)
```

```python
# Print the accuracy score
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy}")

# Print the classification report
print("Classification Report:\n", classification_report(Y_test, Y_pred))
```
